```python
import numpy as np
import matplotlib.pyplot as plt

countries = ['China', 'United States', 'India', 'Indonesia', 'Philippines']
factors = ['amount', 'forest', 'co2']

def min_max_scaler(arr):
    result = []
    for x in arr:
        x = float(x - np.min(arr))/(np.max(arr)- np.min(arr))
        result.append(x)
    return result
        
for i in range(len(countries)):
    plt.figure(figsize=(6,4),dpi = 100)
    plt.title = (countries[i])
    for factor in factors:
        plt.xlabel = ('Year')
        path = 'factor/'+countries[i]+'_'+factor+'.csv'
        file = np.loadtxt(path, delimiter = ',')
        x = file[:,0]
        y = min_max_scaler(file[:,1])
        plt.plot(x, y, label = factor ,marker='o', markersize=3)
        plt.legend()
```
